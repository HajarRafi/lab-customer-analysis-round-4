```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('marketing_customer_analysis.csv')
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
col=[]

for i in range(len(data.columns)):
    col.append(data.columns[i].lower().replace(' ','_'))
data.columns=col
data=data.rename(columns={'employmentstatus':'employment_status'})
```

```python
data
```

```python
data_num = pd.DataFrame(data.select_dtypes(include=np.number))
data_num
```

```python
data_cat = pd.DataFrame(data.select_dtypes(include=object))
data_cat
```

```python
#Use seaborn library to construct distribution plots for the numerical variables
num_col = list(data_num.columns)
for i in num_col:
    sns.displot(data_num[i], bins=40)
    plt.show()
```

```python
#Use Matplotlib to construct histograms
plt.hist(data_cat['education'], bins=40)
plt.show()  

plt.hist(data_cat['marital_status'], bins=40)
plt.show()  
```

```python
correlations_matrix = data_num.corr()
correlations_matrix
```

```python
sns.heatmap(round(correlations_matrix, 2), annot=True)
plt.show()
```

```python
import statsmodels.api as sm
from patsy import dmatrices 
```
